```python
import polars as pl

#считываем датасет и выводим информацию
df = pl.read_csv("train.csv")
print(df)
#пассажиры разных классов
print(df["Pclass"].value_counts())
#кол-во выживших мужчин и женщин
print(df.group_by("Sex").agg((pl.col("Survived")).sum()))
#пассажиры возраст которых более 44
print(df.filter(df["Age"] > 44))
```

```python
import pandas as pd
import bottleneck as bn
import numpy as np
df = pd.read_csv('train.csv')
#cредний возраст
bn.nanmedian(df['Age'])
```

```python
#Стандартное отклонения возраста
bn.nanstd(df['Age'])
```

```python
#умножьте значение столбца Fare на 1.3 и на 1.1
df['Fare_new'] = np.where(df['Pclass'] < 3, df['Fare'] * 1.3, df['Fare'] * 1.1)
df.to_csv('train_modified.csv', index=False)

print(df.info())
```

```python
df_h = pd.read_csv("Housing.csv")
df_h.info()
```

```python
df_h.head(15)
```

```python
df_h.describe()
```

```python
df_h['price']=df_h['price'].astype(np.int32)
df_h['area']=df_h['area'].astype(np.int16)
df_h['bedrooms']=df_h['bedrooms'].astype(np.int8)
df_h['bathrooms']=df_h['bathrooms'].astype(np.int8)
df_h['stories']=df_h['stories'].astype(np.int8)
df_h['parking']=df_h['parking'].astype(np.int8)
df_h['mainroad']=df_h['mainroad'].astype("category")
df_h['guestroom']=df_h['guestroom'].astype("category")
df_h['basement']=df_h['basement'].astype("category")
df_h['hotwaterheating']=df_h['hotwaterheating'].astype("category")
df_h['airconditioning']=df_h['airconditioning'].astype("category")
df_h['prefarea']=df_h['prefarea'].astype("category")
df_h['furnishingstatus']=df_h['furnishingstatus'].astype("category")
df_h.info()
```

* Значение цены всегда положительное и не превышает значения 13.300.000 можно уложить в Int 32
* Значение площади не превышает 16.200 можно уложить в Int16
* Остальные числовые значения можно уложить в int8 (6, 4, 4, 3)
* Остальные object были заменены на категории, потому что в каждом столбце немного уникальных значений (2, 2, 2, 2, 2, 3)
